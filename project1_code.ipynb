```python
#import all packages needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors
#import data file
ngsim_data = pd.read_csv('CIVE202_Spring2025_GroupE_Project3_NGSIM.csv')
#create the plot
trajectory_number = 1
#Creates subset for our current data set of only the first set of cars. 
data_subset = ngsim_data[ngsim_data['trajectory_number'] == trajectory_number]
#calculates the gap distance
gap_distance = data_subset['leader_position(m)'].values - data_subset['follower_position(m)'].values

#set size of plot
plt.figure(figsize=(10,5))
#plots the graph and sets the labels
plt.plot(data_subset['Time'], gap_distance)
plt.xlabel('Time (seconds)')
plt.ylabel('Gap Distance (m)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from PIL import Image
from IPython.display import display

nhts_data = pd.read_csv('CIVE202_Spring2025_GroupE_Project3_NHTS.csv')
 # importing the national household travel survey
print(nhts_data)
nhts_data.columns

nhts_data = nhts_data.dropna(subset=['vehicles_per_household'])
data_grouped = nhts_data.groupby(['vehicles_per_household','census_division']).size().unstack(fill_value=0)
data_grouped.plot(kind='bar',stacked=True)

plt.xlabel('vehicles_per_household')
plt.ylabel('count')
plt.title('Stacked Bar Chart of Make')
plt.xticks(rotation=90)

plt.figure(figsize=(10,20))

data_grouped.plot(kind='bar',
                 stacked=True, 
                 color=sns.color_palette('viridis', len(data_grouped.columns))) 

plt.legend(title = 'Census Division', bbox_to_anchor = (1,1), loc='upper left')
plt.tight_layout()
plt.show()
```
